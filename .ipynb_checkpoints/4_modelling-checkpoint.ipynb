### Define X and y

```python
# Isolate target data
completed_session = features['session_completed']
y = np.asarray(completed_session)
#check size of y array
y.shape
```

```python
#Remove y from the features dataframe
features.pop('session_completed')
print('session_completed removed')
```

```python
#Encode the categorical variables using dummy encoding
#This encodes categorical data 1-0
features_dummy = pd.get_dummies(features)
```

Predictors care about the relative size of different features even though those scales might be arbitrary. StandardScaler fixes this by normalizing each feature to a range of around 1.0 to -1.0 thereby aiming to prevent models from misbehaving. 

```python
#change the dataframe to a matrix of floats so that scaler can be applied
X = features_dummy.as_matrix().astype(np.float)

# Apply the standard scaler
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
#Check the size of the array
print("Dimensions of the feature array:")
print(X.shape)
print
print("Dimensions of the class array:")
print(y.shape)
```

## Fix class imbalance
Nearly 80% of the users completed the first session creating a class imbalance. To addres this oversampling was done on the class 'didn't complete first session' (i.e. 0) using Synthetic Minority Over-sampling technique (SMOTE). More information is available at: http://contrib.scikit-learn.org/imbalanced-learn/generated/imblearn.over_sampling.SMOTE.html

```python
pd.Series(list(y)).value_counts(normalize=True)
```

```python
# Apply regular Synthetic Minority Over-sampling technique (SMOTE)
sm = SMOTE(kind='regular')
X_resampled, y_resampled = sm.fit_sample(X, y)
```

```python
#we can see now that we have acheived balanced classes
print ('before smote:')
print(pd.Series(list(y)).value_counts())
print('after smote')
pd.Series(list(y_resampled)).value_counts()
```

Now lets visualize the impact of SMOTE using a before and after PCA. We can see that the minority classifier feature set in blue has been filled out.

```python
#Lets visualize the impact of SMOTE
visualize_smote(X,y,X_resampled, y_resampled)
```

## Selecting a classifier

Let's compare several fairly unique algorithms: logistic regression, gradient boosting classifier, support vector machines, random forest, and k-nearest-neighbors. I just pass each algorithm to cross validation to determine how often the classifier predicted the correct class.

### Accuracy of classifiers

The accuracy of each classifier was calculated using <a href="http://scikit-learn.org/stable/modules/generated/sklearn.model_selection.KFold.html#sklearn.model_selection.KFold" target="_blank">sklearn K-Folds cross-validator </a>

```python
print('Logistic Regression:')
print("%.3f" % accuracy(y_resampled, run_cv(X_resampled,y_resampled,RF)))
print("Gradient Boosting Classifier:")
print("%.3f" % accuracy(y_resampled, run_cv(X_resampled,y_resampled,GBC)))
print("Random forest:")
print("%.3f" % accuracy(y_resampled, run_cv(X_resampled,y_resampled,RF)))
print("K-nearest-neighbors:")
print("%.3f" % accuracy(y_resampled, run_cv(X_resampled,y_resampled,KNN)))
```

We can see that the Gradient Boosting classifier performs the best. However, the random forest model is nearly as good and tends to be more transparent in terms of feature interactions. Lets take a look at their performance in more detail by looking at the confusion matrices for each classifier.

### Confusion matrices for classifiers

```python
#rename the resampled arrays for simplicity
X = X_resampled
y = y_resampled 

y = np.array(y)
class_names = np.unique(y)

confusion_matrices = [
    ( "Random Forest", confusion_matrix(y,run_cv(X,y,RF)) ),
    ( "K-Nearest-Neighbors", confusion_matrix(y,run_cv(X,y,KNN)) ),
    ( "Gradient Boosting Classifier", confusion_matrix(y,run_cv(X,y,GBC)) ),
    ( "Logisitic Regression", confusion_matrix(y,run_cv(X,y,LR)) )
]

%matplotlib inline

draw_confusion_matrices(confusion_matrices,class_names)
```

In order to gain insight into which features are predictive of first session completion I undertook a random forest analysis and found the the feature importance to be:

Age is the most important feature in determining whether a user will complete the first session or not. For an analysis of the age distribution see feature engineer notebook (ipynb3). It is however not the most correlated to session completion. The symbol matching task and  and android sign up form are more highly correlated to session completion than age.

```python
#This function creates a dataframe with feature importance in descending order, SD, and correlation 
#to model target variable (i.e. 'completed first session')
importance_df = feature_importance_df(final_features.copy())

#Give the feature importance directionality reflecting the direction of the correlation between the feature and 
#the target variable (session completion)
importance_df['plot_value'] = [-1 if x <0 else 1 for x in importance_df['corr_to_session_compl']]
importance_df['plot'] = importance_df['feature_importance']*importance_df['plot_value']

#pull out the top 20 most important features
top_20 = importance_df[['features','plot','standard_error']][:20]

#plot the top 20 most important feature variables
ax = sns.barplot(
    y='features', 
    x=top_20['plot'], 
    data=top_20, 
    label='Label',
    ci=95,
    palette='Blues_r'
    );

ax.set(xlabel='Feature importance', ylabel='Feature names')
ax.figure.tight_layout()
ax.figure.savefig('x_figures/4_modelling/feature_importance15.jpg',dpi =300)
```
